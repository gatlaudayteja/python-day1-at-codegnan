```python
#set functions
#adding elemnts to set
#add(ele): it add the elment to set if elemnts is not in set otherwise nothing
#update(iteable): it add multiple elements in iterator to set
# #s1 = {1,2,3,4,''}

s1 = {1,2,3,4}
s1.add(10)
s1.add(1)
print(s1)
s1.update([20,30,40])
print(s1)
```

```python
s1 = {1,2,3,4}
s1.remove(1)
s1.remove(3)
print(s1)
s1.update([20,30,40])
print(s1)
```

```python
#union
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.union(s2))
#print(s1 / s2) #s1 | s2
```

```python
#intersection
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.intersection(s2))
```

```python
#difference
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.difference(s2))
```

```python
#symmetric difference
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.symmetric_difference(s2))
```

```python
#subset
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.issubset(s2))
```

```python
#superset
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.issubset(s2))
```

```python
#disjoint
s1 = {1,2,3,4}
s2 = {3,4,5,6}
print(s1.isdisjoint(s2))
```

```python
s = set()

# Add element
def add_element(x):
    s.add(x)
    print(" Element added")

# Delete element
def delete_element(x):
    if x in s:
        s.remove(x)
        print(" Element removed")
    else:
        print(" Element not found")

# Total elements
def total_elements():
    print(" Total elements:", len(s))

# Display all elements
def all_elements():
    if not s:
        print(" Set is empty")
    else:
        print(" Set elements:", s)

# Main program
if __name__ == "__main__":
    print("Set Operations Program")

    while True:
        print("""
1. Add element
2. Delete element
3. Total elements
4. Show all elements
5. Exit
""")

        choice = input("Enter choice (1-5): ")

        if choice == "1":
            x = int(input("Enter element: "))
            add_element(x)

        elif choice == "2":
            x = int(input("Enter element to delete: "))
            delete_element(x)

        elif choice == "3":
            total_elements()

        elif choice == "4":
            all_elements()

        elif choice == "5":
            print("üëã Program ended")
            break

        else:
            print("‚ùå Invalid choice")
```
