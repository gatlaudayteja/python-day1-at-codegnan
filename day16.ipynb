```python
#contact book
#1>new contact
#update moblie no
#delete contact
#get contact no 
#print all contact(name,no)

ph_book={}
def add_contact(name:str,number:int):
    if name not in ph_book:           #check name is exists in ph_book or not
        ph_book[name] = number
        print("contact saved")
    else:
        print("Name Already exists")

#update number
def update_number(name:str, number:int):
     if name not in ph_book:           #check name is exists in ph_book or not
         print("contact not exists")
     else:
         ph_book.update((name, number))
         print("moblie number update")

    #delete contact

def delete_contact(name:str):
    if name in ph_book:
        ph_book.pop(name)
        print("successfully deleted")
    else:
        print("contact not found")

    #get contact no

def get_number(name:str):
    number = ph_book.get(name, 'name not found')
    print('the contact number update is:', number)

    #print all contact
def all_contact():
    if not ph_book:
        print("no contact found")
        return
    for name, number in ph_book.items():
        print(f"{name}:{number}")

    # main
if __name__ == '__main__':
    print("welcome to the contact book")
    while True:
        print("\n------ welcome to contact book ------")
        print("1. New Contact")
        print("2. Update Mobile Number")
        print("3. Delete Contact")
        print("4. Get Contact Number")
        print("5. Print All Contacts")
        print("6. Exit")
        print("------------------")


        choice = input("Enter your choice (1-6): ")

        if choice == "1":
            name = input("Enter name: ")
            number = int(input("Enter mobile number: "))
            add_contact(name, number)

        elif choice == "2":
            name = input("Enter name: ")
            number = int(input("Enter new mobile number: "))
            update_number(name, number)

        elif choice == "3":
            name = input("Enter name: ")
            delete_contact(name)

        elif choice == "4":
            name = input("Enter name: ")
            get_number(name)

        elif choice == "5":
            all_contact()

        elif choice == "6":
            print("Exiting Contact Book. Goodbye!")
            break

        else:
            print("Invalid choice. Try again.")



    

```
