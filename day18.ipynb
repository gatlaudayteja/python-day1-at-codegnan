```python
f = open('sample.txt', 'w')
string = '''hello world
               how are you
               iam fine'''
f.write(string)
f.close()
```

```python
f = open('sample.txt', 'w')
string = '''pyhhton is a programming language'''
f.write(string)
f.close()
print("file written successfully")
```

```python
f = open('sample.txt', 'a+')
string = '''java programming language'''
f.write(string)
f.close()
print("file written successfully")
```

```python
f = open('sample.txt', 'r+')
content = f.read()
print(content)
f.close()
print("content added successfully")
```

```python
f = open('sample.txt', 'r+')
content = f.readlines()
print(content[:11])
f.close()
print("content added successfully")
```
